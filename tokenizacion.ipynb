```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
text = "The cat is black and the dog is white".lower()
not_leaked_text = "The cat is black and the dog is white"
```

```python
tokens = word_tokenize(text)
print(tokens)
```

```python
leaked_text = [word for word in tokens if not word in stop_words]
print(leaked_text)
print(not_leaked_text)
```
