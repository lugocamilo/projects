```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_funct(x):
    return x**2 - 5*x + 2
def derivada(x):
    return 2*x -5 #para llegar a 0 toca igualarla a 0
```

```python
x = np.linspace(-20, 20, 100)
y = my_funct(x)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.plot(2.5, my_funct(2.5), 'bo')
```

```python
init_x = 15
init_y = my_funct(init_x)
dx = derivada(init_x)
px = init_x - dx
py = my_funct(px)
```

```python
plt.plot(x,y)
plt.plot(init_x, init_y, 'bo')
plt.plot(px, py, 'bo')
```

```python
lr = 0.1 #learning rate
init_x = 15
init_y = my_funct(init_x)
dx = derivada(init_x)
px = init_x - dx * lr
py = my_funct(px)
```

```python
plt.plot(x,y) #tiene que llegar a 0
plt.plot(init_x, init_y, 'bo')
plt.plot(px, py, 'bo')
for i in range(30):
    px = px - derivada(px) * lr
    py = my_funct(px)
    plt.plot(px, py, 'bo')
    print(px)
```
