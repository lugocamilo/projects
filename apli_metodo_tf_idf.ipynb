```python
import pandas as pd

df = pd.read_csv("C:\\Users\\lugoc\\OneDrive\\Desktop\\NPL - IA\\df\\movie_metadata.csv",
                 encoding="utf-8")
df.head(5)
```

```python
df.drop(columns='color', inplace=True)
df
```

```python
df['genres'] = df['genres'].str.lower().str.split('|').str.join(' ')
df['plot_keywords'] = df['plot_keywords'].str.lower().str.split('|').str.join(' ')
df.head(5)
```

```python
df['text'] = df[['genres', 'plot_keywords']].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
```

```python
df
```

```python
row = df[['genres', 'plot_keywords', 'text']].iloc[0]
print(row)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

tfidf = TfidfVectorizer(max_features=2000)
```

```python
X= tfidf.fit_transform(df['text'])
X
```

```python
peliculas = pd.Series(df.index, index = df['movie_title'])
peliculas
```

```python
peliculas.index = peliculas.index.str.strip()
```

```python
indice = peliculas['Star Wars: Episode VII - The Force Awakens']
```

```python
indice
```

```python
consulta = X[indice]
consulta
```

```python
print(consulta)
```

```python
similitud = cosine_similarity(consulta, X)
```
