```python
import pandas as pd
import gensim 

vectores = gensim.models.KeyedVectors.load_word2vec_format('C:\\Users\\lugoc\\OneDrive\\Desktop\\NPL - IA\\df\\SBW-vectors-300-min5.txt')
```

```python
def analogias(v1, v2, v3):
    similitud = vectores.most_similar(positive=[v1, v3], negative=[v2])
    print(f"{v1} es a {v2} como {similitud[0][0]} es a {v3}")
```

```python
analogias ('rey', 'hombre', 'mujer')
```

```python
analogias('alemania', 'aleman', 'español')
```

```python
analogias('pan', 'trigo', 'leche')
```

```python
analogias('vaca', 'leche', 'huevo')
```

```python
analogias('chimenea', 'humo', 'agua')
```

```python
def cercanos(v):
    vecinos = vectores.most_similar(positive=[v]) #los vecinos mas cercanos del vector que vamos a darle
    print(f"vecinos de: {v}")
    for word, score in vecinos:
        print("\t%s" % word)
```

```python
cercanos('rey')
```

```python
cercanos('infiel')
```
