```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\lugoc\\OneDrive\\Desktop\\ciencia de datos\\datasets\\censo_hogares.csv',
                 sep='|', #-----> Separador o delimitador
                 encoding= 'UTF-8') #-------> Diccionario de caracteres UTF-8 o latin1
df.head(5)
```

```python
df.describe()
```

```python
df.replace('?', np.nan, inplace = True)
```

```python
df.isnull().sum()
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
#Eliminamos columna de numero_fallecidos

df.drop(['numero_fallecidos'], axis=1, inplace = True)
df.head(5)
```

```python
#Lista para eliminar los datos de las columnas con null
columnas_vit = ['numero_cuartos', 'numero_dormitorios', 'total_personas', 'orden_hogar']
#Borramos las filas donde en cualquiera de las columnas hayan nulos
df.dropna(subset=columnas_vit, inplace = True)
#verificar que el contador de nulos para las columnas sea cero
print("Nulos restantes en columnas vitales:")
print(df[columnas_vit].isnull().sum())

#imprimimos cuanto quedo el dataset
print(f"Registros totales despues de la limpieza: {len(df):,}")
```

```python
media_personas = df['total_personas'].mean()
mediana_personas = df['total_personas'].median()
moda_personas = df['total_personas'].mode()[0]

print(f"La media del total de personas es: {media_personas:.2f}")
print(f"La mediana del total de personas es: {mediana_personas}")
print(f"La moda de el total de personas es: {moda_personas}")
```

```python
#creacion de la tabla de frecuencia

tabla_fre_personas = df['total_personas'].value_counts(normalize=True).sort_index() * 100

#Mostrar el resultado de la tabla de frecuencia 
print(f"Porcentaje de hogares por numero de personas: {tabla_fre_personas.head(5)}")
```

```python
#asegurar que los datos solo sean numeros
cols_num = ['numero_cuartos','numero_dormitorios','total_personas']
for col in cols_num:
    df[col]= pd.to_numeric(df[col], errors= 'coerce')
    
#Limpiamos los ceros en los dormitorios 
df = df[df['numero_dormitorios'] > 0]

#Se crea una metrica : personas por dormitorio
df['per_por_dormitorio'] =df['total_personas'] / df['numero_dormitorios']

#Se hace la clasificacion 
df['hacinamiento critico'] = df['per_por_dormitorio'] > 3

print("Metrica de hacinamiento ")
print(df[['total_personas', 'numero_dormitorios', 'per_por_dormitorio']].head())
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,6))

#Calculamos la frecuencia del porcentaje
datos_hogar =df['total_personas'].value_counts(normalize=True).sort_index() * 100

#Creamos la grafica 
sns.barplot(x=datos_hogar.index[:6], y=datos_hogar.values[:6], palette="viridis" )

plt.title('Distribucion del tamaño de hogares (censo 2018)', fontsize =15)
plt.xlabel('Numero de personas en el hogar ', fontsize = 12)
plt.ylabel('Porcentaje de hogares ', fontsize=12)
plt.ylim(0,30)

for i, v in enumerate(datos_hogar.values[:6]):
    plt.text(i, v + 0.5, f"{v:.1f}%", ha='center', fontweight='bold')
plt.show()
```

```python
#Heatmap
#Filas= Peronas, columnas =cuartos

heatmap_data = pd.crosstab(df['total_personas'], df['numero_cuartos'])
heatmap_data = heatmap_data.iloc[:7, :7]

plt.figure(figsize=(12,8))
sns.heatmap(heatmap_data, annot= True, fmt="d", cmap="YlGnBu", cbar_kws={'label' : 'cantidad de hogares'})

plt.title("Relacion: Total de personas vs numero de cuartos")
plt.xlabel("Numero de cuartos", fontsize = 12)
plt.ylabel("Total de personas en el hogar", fontsize=12)
plt.show()
```

```python
#grafica geografica
plt.figure(figsize=(10,6))

depto_mean = df.groupby('nombre_departamento')['total_personas'].mean().sort_values()

sns.barplot(x=depto_mean.values, y=depto_mean.index, palette="magma" )

plt.title('Promedio de Personas por Hogar según Departamento', fontsize=15)
plt.xlabel('Promedio de Personas', fontsize=12)
plt.ylabel('Departamento', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
